````python
import itertools
import math
from functools import reduce, wraps
import matplotlib.pyplot as plt

# --- Генераторы ---
def gen_rectangle(width=1, height=1, offset=(0, 0), step=(2, 0)):
    """
    Генерирует бесконечную последовательность непересекающихся прямоугольников
    заданной ширины и высоты, начиная с точки offset и смещаясь на вектор step.
    """
    x0, y0 = offset
    dx, dy = step
    while True:
        yield ((x0, y0), (x0 + width, y0), (x0 + width, y0 + height), (x0, y0 + height))
        x0 += dx; y0 += dy # Смещаем начальную точку для следующего прямоугольника


def gen_triangle(side=1, offset=(0, 0), step=(2, 0)):
    """
    Генерирует бесконечную последовательность правильных треугольников
    со стороной side, начиная с offset и смещаясь на step.
    """
    x0, y0 = offset
    dx, dy = step
    h = side * math.sqrt(3) / 2
    while True:
        yield ((x0, y0), (x0 + side, y0), (x0 + side/2, y0 + h))
        x0 += dx; y0 += dy


def gen_hexagon(side=1, offset=(0, 0), step=(3, 0)):
    """
    Генерирует бесконечную последовательность правильных шестиугольников
    со стороной side, начиная с offset и смещаясь на step.
    """
    x0, y0 = offset
    dx, dy = step
    angles = [math.radians(a) for a in range(0, 360, 60)]
    while True:
      # Вычисляем 6 вершин через тригонометрические функции
        yield tuple((x0 + side * math.cos(a), y0 + side * math.sin(a)) for a in angles)
        x0 += dx; y0 += dy # Смещаем центр для следующего шестиугольника

# --- Преобразования ---
def tr_translate(poly, tx, ty, origin=(0,0)):
    """Параллельный перенос многоугольника на вектор (tx, ty)."""
    return tuple((x + tx, y + ty) for x, y in poly) # создаём новый неизменяемый кортеж


def tr_rotate(poly, angle, origin=(0,0)):
    """Поворот многоугольника на угол angle (градусы) вокруг origin."""
    ox, oy = origin
    rad = math.radians(angle)  # конвертирует градусы в радианы
    c, s = math.cos(rad), math.sin(rad)
    return tuple((ox + (x-ox)*c - (y-oy)*s,
                  oy + (x-ox)*s + (y-oy)*c) for x,y in poly)


def tr_symmetry(poly, axis='x', origin=(0,0)):
    """Отражение многоугольника относительно оси 'x' или 'y'."""
    ox, oy = origin
    if axis == 'x': return tuple((x, 2*oy - y) for x,y in poly) # y-координата "зеркалится" относительно горизонтальной линии y=oy
    if axis == 'y': return tuple((2*ox - x, y) for x,y in poly)
    raise ValueError("Ось должна быть 'x' или 'y'") # Срабатывает, если axis содержит недопустимое значение


def tr_homothety(poly, factor, origin=(0,0)):
    """Гомотетия (масштабирование) многоугольника относительно origin."""
    ox, oy = origin
    return tuple((ox + factor*(x-ox), oy + factor*(y-oy)) for x,y in poly)

# --- Утилиты ---
def polygon_area(poly):
    """Вычисление площади многоугольника по формуле Гаусса (шнуровки)."""
    return abs(sum(
        # Формула состоит из двух частей:
        # 1. Сумма произведений x_i * y_i+1 (диагонали ↘)
        poly[i][0] * poly[(i+1)%len(poly)][1] -

        # 2. Сумма произведений y_i * x_i+1 (диагонали ↙)
        poly[(i+1)%len(poly)][0] * poly[i][1]

        # Цикл по всем вершинам многоугольника:
        for i in range(len(poly))

    ) / 2)  # Абсолютное значение и деление на 2

def side_lengths(poly):
    """Длины сторон многоугольника."""
    return [math.hypot(poly[(i+1)%len(poly)][0] - poly[i][0], poly[(i+1)%len(poly)][1] - poly[i][1])
            for i in range(len(poly))]


def point_in_poly(pt, poly):
    """Определяет, находится ли точка внутри полигона методом ray casting.

    Алгоритм: выпускает горизонтальный луч из точки и считает пересечения с границей

    Параметры:
        pt: (x, y) - координаты проверяемой точки
        poly: [(x1,y1), (x2,y2), ...] - вершины полигона (замкнутый контур)

    Возвращает:
        True если точка внутри, False если снаружи
    """
    # Распаковываем координаты точки
    x, y = pt

    #  Инициализируем флаг
    inside = False

    # Шаг  Создаем пары соседних вершин для обработки ребер
    for (x1,y1), (x2,y2) in zip(poly, poly[1:]+poly[:1]):

        # Проверка условия пересечения луча с ребром
        ray_crosses_segment = (y1 > y) != (y2 > y)

        # Точка находится слева от точки пересечения луча с ребром
        x_intersect = (x2 - x1) * (y - y1) / (y2 - y1) + x1
        point_is_left = x < x_intersect

        # Если оба условия выполнены - луч пересекает ребро
        if ray_crosses_segment and point_is_left:
            inside = not inside  # Меняем состояние


    return inside

# --- Фильтры ---
def _cross(o, a, b):
    return (a[0]-o[0])*(b[1]-o[1]) - (a[1]-o[1])*(b[0]-o[0])


def flt_convex_polygon(poly):
    """Проверяет, является ли полигон выпуклым.

    Алгоритм:
    1. Вычисляет векторные произведения для последовательных вершин
    2. Проверяет одинаковость знаков всех произведений

    Параметры:
        poly: [(x1,y1), (x2,y2), ...] - вершины полигона в порядке обхода

    Возвращает:
        True если полигон выпуклый, False если вогнутый
    """

    signs = [
        # Для каждой тройки последовательных вершин (A,B,C) вычисляем векторное произведение AB × BC (через вспомогательную _cross)
        _cross(poly[i], poly[(i+1)%len(poly)], poly[(i+2)%len(poly)])

        # Перебираем все вершины полигона
        for i in range(len(poly))
    ]

    #Проверяем условие выпуклости все произведения должны быть одного знака (все ≥0 или все ≤0)
    return all(s >= 0 for s in signs) or all(s <= 0 for s in signs)

def flt_angle_point(poly, point):
    """True, если один из углов совпадает с точкой point."""
    return point in poly


def flt_area_less(poly, max_area):
    """True, если площадь полигона меньше max_area."""
    return polygon_area(poly) < max_area


def flt_short_side(poly, min_length):
    """True, если самая короткая сторона меньше min_length.
    Алгоритм:
        1. Вычисляет длины всех сторон через side_lengths()
        2. Находит минимальную длину
        3. Сравнивает с min_length
        """
    return min(side_lengths(poly)) < min_length


def flt_point_inside(poly, point):
    """True, если выпуклый полигон содержит точку point внутри."""
    return flt_convex_polygon(poly) and point_in_poly(point, poly)


def flt_polygon_angles_inside(poly, other):
    """True, если выпуклый полигон содержит хотя бы один угол другого полигона other."""
    return flt_convex_polygon(poly) and any(point_in_poly(p, poly) for p in other)

# Необязательный: фильтр между двумя прямыми

def flt_between_lines(poly, m1, m2):
    """True, если все вершины лежат между прямыми y=m1*x и y=m2*x."""
    for x,y in poly: # Перебираем все вершины полигона
        if x == 0:  # Для вертикальных точек (x=0) пропускаем проверку
           continue
         # Вычисляем угловой коэффициент для текущей точки
        slope = y / x
        # Проверяем нахождение между двумя прямыми
        if not (min(m1, m2) <= slope <= max(m1, m2)):
            return False
    return True

# Декораторы
def decorator_transform(op):
    """Фабрика декораторов для применения преобразований к элементам последовательности.

    Параметры:
        op: функция-преобразование, которая будет применена к каждому элементу

    Возвращает:
        Декоратор, который преобразует входную последовательность
    """
    def deco(func):
        """Собственно декоратор, который будет оборачивать целевую функцию.

        Параметры:
            func: целевая функция, которую нужно декорировать

        Сохраняет метаданные оригинальной функции с помощью @wraps
        """
        @wraps(func)  # Сохраняет имя и docstring оригинальной функции
        def wrapper(seq, *args, **kwargs):
            """Обертка, которая выполняет преобразование перед вызовом функции.

            1. Применяет op к каждому элементу seq (через map)
            2. Передает преобразованную последовательность в func
            """
            transformed_seq = map(op, seq)  # Ленивое преобразование
            return func(transformed_seq, *args, **kwargs)
        return wrapper
    return deco


def decorator_filter(flt):
    """Фабрика декораторов для фильтрации последовательностей.

    Параметры:
        flt: функция-фильтр (возвращает bool), применяемая к каждому элементу

    Возвращает:
        Декоратор, который фильтрует входную последовательность
    """
    def deco(func):
        """Создаваемый декоратор, сохраняющий метаданные функции."""
        @wraps(func)  # Сохраняет имя, docstring и другие метаданные func
        def wrapper(seq, *args, **kwargs):
            """Обертка, которая:
            1. Фильтрует последовательность с помощью flt
            2. Передает отфильтрованную последовательность в func
            """
            filtered_seq = filter(lambda p: flt(p), seq)  # Ленивая фильтрация
            return func(filtered_seq, *args, **kwargs)
        return wrapper
    return deco

# Алиасы для декораторов

dec_translate = decorator_transform(lambda p: tr_translate(p,1,1)) # Сдвигает все полигоны на вектор (1,1)
dec_rotate    = decorator_transform(lambda p: tr_rotate(p,45)) # Поворачивает полигоны на 45° вокруг начала координат
dec_symmetry  = decorator_transform(lambda p: tr_symmetry(p,'x')) # Отражает полигоны по оси X
dec_homothety = decorator_transform(lambda p: tr_homothety(p,2)) # Масштабирует полигоны в 2 раза относительно (0,0)

dec_flt_convex = decorator_filter(flt_convex_polygon) # Оставляет только выпуклые полигоны
dec_flt_angle = decorator_filter(lambda p: flt_angle_point(p,(0,0))) # Полигоны, имеющие вершину в точке (0,0)
dec_flt_area_less = decorator_filter(lambda p: flt_area_less(p,1.0)) # Полигоны площадью < 1.0
dec_flt_short_side = decorator_filter(lambda p: flt_short_side(p,1.0)) # Полигоны с хотя бы одной стороной < 1.0
dec_flt_point_inside = decorator_filter(lambda p: flt_point_inside(p,(0,0))) # Полигоны, содержащие точку (0,0) внутри
dec_flt_angles_inside = decorator_filter(lambda p: flt_polygon_angles_inside(p,((0,0),(1,0),(0,1)))) # Полигоны, содержащие хотя бы одну вершину из треугольника (0,0)-(1,0)-(0,1)

# --- Агрегаторы ---
def agr_origin_nearest(polys):
    """Возвращает полигон, ближайший к началу координат."""
    return min(polys, key=lambda poly: min(math.hypot(x,y) for x,y in poly))


def agr_max_side(polys):
    """Возвращает полигон с наибольшей длиной стороны."""
    return max(polys, key=lambda poly: max(side_lengths(poly)))


def agr_min_area(polys):
    """Возвращает полигон с наименьшей площадью."""
    return min(polys, key=lambda poly: polygon_area(poly))


def agr_perimeter(polys):
    """Вычисляет суммарный периметр всех полигонов."""
    return sum(sum(side_lengths(poly)) for poly in polys)


def agr_area(polys):
    """Вычисляет суммарную площадь всех полигонов."""
    return sum(polygon_area(poly) for poly in polys)

# --- Соединение и счёт ---
def zip_polygons(*iters):
    """Склеивает несколько последовательностей полигонов, объединяя вершины."""
    for polys in zip(*iters):
        yield tuple(pt for poly in polys for pt in poly)


def count_2D(start, step):
    """Генератор бесконечной последовательности 2D точек с арифметической прогрессией.

    Параметры:
        start: (x0, y0) - начальная точка
        step: (sx, sy) - шаг приращения по x и y

    Возвращает:
        Бесконечный генератор точек вида (x0 + i*sx, y0 + i*sy)
        где i = 0, 1,  ...

    """
    # Распаковываем начальные координаты
    x0, y0 = start

    # Распаковываем шаги приращения
    sx, sy = step

    # Счетчик итераций (начинается с 0)
    i = 0

    # Бесконечный цикл генерации
    while True:
        # Вычисляем текущую точку
        current_point = (x0 + i*sx, y0 + i*sy)

        # Возвращаем точку через yield
        yield current_point

        # Увеличиваем счетчик
        i += 1


def zip_tuple(*iters):
    """Zip-объединение нескольких последовательностей точек."""
    yield from zip(*iters)

# --- Визуализация ---
def visualize(polys, figsize=(6,6), title=None):
    """
    Визуализирует набор полигонов с помощью matplotlib.

    Параметры:
        polys: итерируемый объект с полигонами (каждый полигон - кортеж точек (x,y))
        figsize: кортеж (ширина, высота) графика в дюймах (по умолчанию 6x6)
        title: строка - заголовок графика (опционально)
    """
    # 1. Создание фигуры и осей с заданным размером
    fig, ax = plt.subplots(figsize=figsize)

    # 2. Отрисовка каждого полигона
    for poly in polys:
        # 2.1. Замыкаем полигон, добавляя первую точку в конец
        closed_poly = poly + (poly[0],)

        # 2.2. Разделяем координаты x и y
        xs, ys = zip(*closed_poly)

        # 2.3. Отрисовываем полигон линией
        ax.plot(xs, ys)

    # 3. Настройка внешнего вида
    ax.set_aspect('equal')  # Одинаковый масштаб по осям

    # 4. Добавление заголовка при наличии
    if title:
        ax.set_title(title)

    # 5. Отображение графика
    plt.show()

# --- Демонстрация и тесты ---
if __name__ == '__main__':
    # 1. Семь смешанных фигур
    mixed = itertools.islice(
        itertools.chain(gen_rectangle(), gen_triangle(step=(0,2)), gen_hexagon(step=(2,2))),
        7)
    visualize(list(mixed), title='7 смешанных фигур')

    # 2. Три ленты под углом
    angles = [30, 0, -30]
    gens = [gen_rectangle(step=(2,0)), gen_triangle(step=(2,0)), gen_hexagon(step=(2,0))]
    for angle, gen, name in zip(angles, gens, ['прямоуг.', 'треуг.', 'шестиг.']):
        seq = itertools.islice(map(lambda p: tr_rotate(p, angle), gen), 7)
        visualize(list(seq), title=f'Лента {name} под {angle}°')

    # 3. Пересекающиеся ленты
    r1 = itertools.islice(map(lambda p: tr_translate(p,1,2), gen_rectangle()),7)
    r2 = itertools.islice(map(lambda p: tr_translate(p,2,1), gen_rectangle()),7)
    visualize(list(r1) + list(r2), title='Пересекающиеся ленты')

    # 4. Симметричные треугольники
    t1 = itertools.islice(map(lambda p: tr_symmetry(p,'x'), gen_triangle(step=(2,0))),7)
    t2 = itertools.islice(map(lambda p: tr_symmetry(p,'y'), gen_triangle(step=(2,0))),7)
    visualize(list(t1) + list(t2), title='Симметричные ленты треугольников')

    # 5. Квадраты в разном масштабе между прямыми y=0.5x и y=2x
    base_quad = ((0.5, 1), (1, 1), (1, 2), (0.5, 2))
    quads = [tr_homothety(base_quad, 1 + 0.3*i) for i in range(7)]
    visualize(quads, title='Квадраты в разном масштабе между прямыми')

    # 6. Фильтры
    mix = itertools.chain(gen_rectangle(), gen_triangle(), gen_hexagon())
    conv6 = itertools.islice(filter(flt_convex_polygon, mix), 6)
    visualize(list(conv6), title='6 выпуклых многоугольников')

    mix2 = itertools.chain(gen_triangle(step=(1,1)), gen_hexagon(step=(1,1)))
    short4 = itertools.islice(filter(lambda p: flt_short_side(p,1), mix2), 4)
    visualize(list(short4), title='≤4 фигур с короткой стороной <1')

    mix15 = list(itertools.islice(itertools.chain(gen_rectangle(step=(1,1)), gen_hexagon(step=(1,1))), 15))
    inters = [p for p in mix15 if any(set(p) & set(q) for q in mix15 if p is not q)]
    visualize(inters, title='Пересекающиеся из 15')

    # 7. Декораторы
    @dec_translate
    def demo_t(seq): return list(seq)
    @dec_flt_convex
    def demo_f(seq): return list(seq)
    print('Декоратор translate:', demo_t(itertools.islice(gen_rectangle(step=(2,0)), 3)))
    print('Декоратор convex:', demo_f(itertools.islice(gen_triangle(step=(2,0)), 3)))

    # 8. Агрегаторы
    quads5 = list(itertools.islice(gen_rectangle(step=(3,0)), 5))
    print('Ближайший к началу:', agr_origin_nearest(quads5))
    print('Макс. сторона:', agr_max_side(quads5))
    print('Мин. площадь:', agr_min_area(quads5))
    print('Сумма периметров:', agr_perimeter(quads5))
    print('Сумма площадей:', agr_area(quads5))

    # 9. zip и count
    zpol = list(zip_polygons(itertools.islice(gen_triangle(), 2), itertools.islice(gen_hexagon(), 2)))
    print('zip_polygons:', zpol)
    cnt = list(itertools.islice(count_2D((0,0), (1,2)), 5))
    print('count_2D:', cnt)
    zt = list(zip_tuple([(1,1),(2,2),(3,3),(4,4)], [(2,2),(3,3),(4,4),(5,5)], [(3,3),(4,4),(5,5),(6,6)]))
    print('zip_tuple:', zt)
````
